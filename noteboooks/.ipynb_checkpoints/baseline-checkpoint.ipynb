```python
from sklearn import preprocessing
from sklearn import linear_model
from sklearn import metrics
import joblib
import shap

def run(fold):
    df = train_folds
    train_df = df[df.kfold != fold].reset_index(drop=True)
    valid_df = df[df.kfold == fold].reset_index(drop=True)
    
    xtrain = train_df.drop(["id", "target", "kfold"], axis=1)
    ytrain = train_df["target"]
    xvalid = valid_df.drop(["id", "target", "kfold"], axis=1)
    yvalid = valid_df["target"]
    xtest = test.drop(["id"], axis=1)
    
    scaler = preprocessing.StandardScaler()
    xtrain_scl = scaler.fit_transform(xtrain)
    xvalid_scl = scaler.fit_transform(xvalid)
    xtest_scl = scaler.transform(xtest)
    
    model = linear_model.LogisticRegression(
        class_weight="balanced", 
        penalty="l1", 
        C=0.1, 
        solver="liblinear"
    )
    
    model.fit(xtrain_scl, ytrain)
    pred_valid = model.predict(xvalid_scl)
    auc = metrics.roc_auc_score(yvalid, pred_valid)
    
    pred_test = model.predict_proba(xtest_scl)
    pred_lr = pd.DataFrame()
    pred_lr["id"] = test.id
    pred_lr["pred_proba"] = pred_test[:, 1]
    pred_lr.to_csv(f"../input/pred_lr_{fold}.csv", index=False)
    
    print(f"Fold {fold}, AUC={auc:.3f}")
    
    joblib.dump(
        model,
        f"../models/lr_{fold}.bin"
    )
    
    explainer = shap.LinearExplainer(model, xtrain_scl)
    shap_values = explainer.shap_values(xtrain_scl)
    shap.summary_plot(shap_values, xtrain_scl)
    
if __name__ == "__main__":
    for i in range(5):
        run(i)
        
```

```python
from mlxtend import feature_selection
from mlxtend import plotting
import matplotlib.pyplot as plt

def feat_selector(fold):
    df = train_folds
    train_df = df[df.kfold != fold].reset_index(drop=True)
    valid_df = df[df.kfold == fold].reset_index(drop=True)
    
    xtrain = train_df.drop(["id", "target", "kfold"], axis=1)
    ytrain = train_df["target"]
    xvalid = valid_df.drop(["id", "target", "kfold"], axis=1)
    yvalid = valid_df["target"]
    xtest = test.drop(["id"], axis=1)
    
    scaler = preprocessing.StandardScaler()
    xtrain_scl = scaler.fit_transform(xtrain)
    xvalid_scl = scaler.fit_transform(xvalid)
    xtest_scl = scaler.transform(xtest)
    
    model = linear_model.LogisticRegression(
        class_weight="balanced", 
        penalty="l1", 
        C=0.1, 
        solver="liblinear"
    )
    
    feat_selector = feature_selection.SequentialFeatureSelector(
        model,
        k_features=(10, 15),
        forward=True, 
        floating=False, 
        scoring="roc_auc",
        verbose=0,
        cv=0
    )
    
    feat_selector.fit(xtrain_scl, ytrain)
    plotting.plot_sequential_feature_selection(
        feat_selector.get_metric_dict(),
        kind="std_dev"
    )
    
    plt.title("Sequential Forward Selection (w. StdDev)")
    plt.grid()
    plt.show()
    
    top_features = list(feat_selector.k_feature_names_)
    
    return top_features
```

```python
f0 = feat_selector(0)
f1 = feat_selector(1)
f2 = feat_selector(2)
f3 = feat_selector(3)
f4 = feat_selector(4)
```

```python
def lr_top_feature(fold, features):
    df = train_folds
    train_df = df[df.kfold != fold].reset_index(drop=True)
    valid_df = df[df.kfold == fold].reset_index(drop=True)
    
    xtrain = train_df[features]
    ytrain = train_df["target"]
    xvalid = valid_df[features]
    yvalid = valid_df["target"]
    xtest = test[features]
    
    scaler = preprocessing.StandardScaler()
    xtrain_scl = scaler.fit_transform(xtrain)
    xvalid_scl = scaler.fit_transform(xvalid)
    xtest_scl = scaler.transform(xtest)
    
    model = linear_model.LogisticRegression(
        class_weight="balanced", 
        penalty="l1", 
        C=0.1, 
        solver="liblinear"
    )
    
    model.fit(xtrain_scl, ytrain)
    pred_valid = model.predict(xvalid_scl)
    auc = metrics.roc_auc_score(yvalid, pred_valid)
     
    print(f"Fold {fold}, AUC={auc:.3f}")
    
```

```python
features = [f0, f1, f2, f3, f4]
for feat in features:
    for i in range(5):
        lr_top_feature(i, feat)
```

```python
def lr_grid_search(fold):
    df = train_folds
    train_df = df[df.kfold != fold].reset_index(drop=True)
    valid_df = df[df.kfold == fold].reset_index(drop=True)
    
    xtrain = train_df.drop(["id", "target", "kfold"], axis=1)
    ytrain = train_df["target"]
    xvalid = valid_df.drop(["id", "target", "kfold"], axis=1)
    yvalid = valid_df["target"]
    xtest = test.drop(["id"], axis=1)
    
    scaler = preprocessing.StandardScaler()
    xtrain_scl = scaler.fit_transform(xtrain)
    xvalid_scl = scaler.fit_transform(xvalid)
    xtest_scl = scaler.transform(xtest)
    
    clf = linear_model.LogisticRegression(
        solver="liblinear",
        max_iter=1000
    )
    
    param_grid = {
        "class_weight": ["balanced", None], 
        "penalty": ["l2"], 
        "C": [0.001, 0.01, 0.1, 1.0, 10.0, 100.0], 
        "solver": ["newton-cg", "lbfgs", "sag"]
    }
    
    model = model_selection.GridSearchCV(
        estimator=clf,
        param_grid=param_grid,
        scoring="roc_auc",
        cv=5
    )
    
    model.fit(xtrain_scl, ytrain)
    print(f"Best score: {model.best_score_}")
    
    print("Best parameters set:")
    
    best_parameters = model.best_estimator_.get_params()
    for param_name in sorted(param_grid.keys()):
        print(f"\t{param_name}: {best_parameters[param_name]}")
    
if __name__ == "__main__":
    for i in range(5):
        lr_grid_search(i)
```

```python
from sklearn import ensemble

def rf(fold):
    df = train_folds
    train_df = df[df.kfold != fold].reset_index(drop=True)
    valid_df = df[df.kfold == fold].reset_index(drop=True)
    
    xtrain = train_df.drop(["id", "target", "kfold"], axis=1)
    ytrain = train_df["target"]
    xvalid = valid_df.drop(["id", "target", "kfold"], axis=1)
    yvalid = valid_df["target"]
    xtest = test.drop(["id"], axis=1)
    
    scaler = preprocessing.StandardScaler()
    xtrain_scl = scaler.fit_transform(xtrain)
    xvalid_scl = scaler.fit_transform(xvalid)
    xtest_scl = scaler.transform(xtest)
    
    model = ensemble.RandomForestClassifier()
    
    model.fit(xtrain_scl, ytrain)
    pred_valid = model.predict(xvalid_scl)
    auc = metrics.roc_auc_score(yvalid, pred_valid)
        
    print(f"Fold {fold}, AUC={auc:.3f}")
        
if __name__ == "__main__":
    for i in range(5):
        rf(i)
```

```python
def rf_grid_search(fold):
    df = train_folds
    train_df = df[df.kfold != fold].reset_index(drop=True)
    valid_df = df[df.kfold == fold].reset_index(drop=True)
    
    xtrain = train_df.drop(["id", "target", "kfold"], axis=1)
    ytrain = train_df["target"]
    xvalid = valid_df.drop(["id", "target", "kfold"], axis=1)
    yvalid = valid_df["target"]
    xtest = test.drop(["id"], axis=1)
    
    scaler = preprocessing.StandardScaler()
    xtrain_scl = scaler.fit_transform(xtrain)
    xvalid_scl = scaler.fit_transform(xvalid)
    xtest_scl = scaler.transform(xtest)
    
    clf = ensemble.RandomForestClassifier()
    
    param_grid = {
        "n_estimators": [10, 50, 100, 1000],
        "max_depth": [3, 5, 15, None],
    }
    
    model = model_selection.GridSearchCV(
        estimator=clf,
        param_grid=param_grid,
        scoring="roc_auc",
        cv=5
    )
    
    model.fit(xtrain_scl, ytrain)
    
    print(f"Best score: {model.best_score_}")
    
    print("Best parameters set:")
    best_parameters = model.best_estimator_.get_params()
    for param_name in sorted(param_grid.keys()):
        print(f"\t{param_name}: {best_parameters[param_name]}")

        
if __name__ == "__main__":
    for i in range(5):
        rf_grid_search(i)
```

```python
from sklearn import preprocessing
import matplotlib.pyplot as plt

X_train = train.drop(["id", "target"], axis=1)    
y_train = train["target"]
X_test = test.drop(["id"], axis=1)

poly = preprocessing.PolynomialFeatures(2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.fit_transform(X_train)

corr = pd.DataFrame(X_train_poly).corrwith(y_train)

scores = []
for i in range(10, 510, 5):
    top_corr_cols = list(corr.abs().sort_values().tail(i).reset_index()["index"].values)
    X_train_poly1 = X_train_poly[:,top_corr_cols]
    X_test_poly1 = X_train_poly[:,top_corr_cols]
    
    xtrain, xvalid, ytrain, yvalid = model_selection.train_test_split(
        X_train_poly1,
        y_train,
        test_size=0.2,
        random_state=42,
    )
    
    model = linear_model.LogisticRegression()
    model.fit(xtrain, ytrain)
    pred = model.predict(xvalid)
    auc = metrics.roc_auc_score(yvalid, pred)
    #print(f"Features: {i}, AUC= {auc}")
    scores.append(auc)

poly_features = pd.DataFrame()
poly_features["n_features"] = list(range(10, 510, 5))
poly_features["auc"] = scores

plt.subplots(figsize=(10, 5))
sns.lineplot(
    data=poly_features, 
    x="n_features",
    y="auc",
);
plt.show()
```
